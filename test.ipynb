```python
import numpy as np
def ff(d):
    br = np.array(list(range(2**(d+1))))
    return br + 2**(d+1)-2

def inverse_ff(element):
    d = int(np.log2(element + 2)) - 1
    return d
```

```python
import torch 
from torch import tensor
import matplotlib.pyplot as plt
```

```python
z = torch.cat([torch.randn((2,500)) + tensor([[2],[0]]), torch.randn((2,500))-tensor([[3],[0]])], dim=1)
```

```python
plt.scatter(*z, c=[0]*500+[1]*500)
```

```python
plt.scatter(*torch.sigmoid(z),c=[0]*500+[1]*500)
```

```python
plt.scatter(*torch.softmax(z, dim=0),c=[0]*500+[1]*500)
```
